```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Ganti 'StudentsPerformance.csv' dengan nama file dataset Anda
df = pd.read_csv('StudentsPerformance.csv')

# Lihat sekilas data
print(df.info())
print(df.head())

# Statistik deskriptif
print(df.describe())

# Lihat distribusi nilai
print(df['gender'].describe())
print(df['race/ethnicity'].describe())
print(df['lunch'].describe())
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['gender'], kde=True, color='blue', label='Math Score', bins=20)
sns.histplot(df['race/ethnicity'], kde=True, color='green', label='Reading Score', bins=20)
sns.histplot(df['lunch'], kde=True, color='red', label='Writing Score', bins=20)
plt.title('Distribusi Nilai Ujian')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='math score', y='math score', palette='Set2')
plt.title('Boxplot Nilai Matematika Berdasarkan Gender')
plt.xlabel('gender')
plt.ylabel('Nilai Matematika')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[['math score', 'reading score', 'writing score']].corr(), annot=True, cmap='coolwarm')
plt.title('Korelasi Antara Nilai Ujian')
plt.show()
```
