```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Ganti 'CarFeaturesDataset.csv' dengan file dataset Anda
df = pd.read_csv('data.csv')

# Lihat informasi dataset
print(df.info())
print(df.head())
print(df.describe(include='all'))  # Melihat statistik deskriptif
print(df.isnull().sum())  # Melihat jumlah nilai hilang pada setiap kolom
for col in df.select_dtypes(include='object').columns:
    print(f"{col}: {df[col].nunique()} unique values")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Engine HP'], kde=True, color='blue', bins=30)
plt.title('Engine HP')
plt.xlabel('Harga')
plt.ylabel('Frekuensi')
plt.show()


```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Engine HP', y='Popularity', palette='Set2')
plt.title('Harga Mobil Berdasarkan Merek')
plt.xlabel('Merek')
plt.ylabel('Harga')
plt.xticks(rotation=45)
plt.show()
```
